```python
import plotly.graph_objects as go
```

```python
datos = [equipo_a, equipo_b]
nombres_equipos = ['Equipo A', 'Equipo B']

fig = go.Figure()

for i in range(len(datos)):
    fig.add_trace(go.Box(
        y=datos[i],
        name=nombres_equipos[i],
        boxpoints='all',
        jitter=0.5,
        marker=dict(
            size=4
        ),
        line=dict(
            width=1
        )
    ))

fig.update_layout(
    title='Boxplot - Comparación de Alturas',
    yaxis=dict(title='Altura (cm)'),
    xaxis=dict(title='Equipos')
)

fig.show()
```

Resumiendo:

Para los datos de los equipos A y B:

*Equipo B:

Media: 180.4212
Desvío estándar: 0.3362

La media es un valor que representa el promedio de los datos de altura en cada equipo. En el caso del equipo A, la media es aproximadamente 179.1064 cm, lo que indica que en promedio los jugadores de este equipo tienen una altura cercana a los 179 cm. Por otro lado, en el equipo B, la media es de aproximadamente 180.4212 cm, lo que sugiere que los jugadores de este equipo tienen una altura promedio ligeramente superior a los 180 cm.


El desvío estándar es una medida de dispersión que indica qué tan dispersos están los datos alrededor de la media. En el equipo A, el desvío estándar es de aproximadamente 27.8839 cm, lo que implica que las alturas de los jugadores en este equipo están más dispersas alrededor de la media. En cambio, en el equipo B, el desvío estándar es de aproximadamente 0.3362 cm, lo que indica que las alturas de los jugadores en este equipo están mucho más cercanas a la media y hay menos variabilidad en las alturas.


En resumen, el equipo A tiene una media ligeramente menor y un desvío estándar más grande, lo que sugiere que sus jugadores tienen una mayor variabilidad en las alturas. Por otro lado, el equipo B tiene una media ligeramente mayor y un desvío estándar más pequeño, lo que indica que sus jugadores tienen alturas más consistentes y menos variabilidad.

**El coeficiente de variación** (CV) es una medida relativa de la variabilidad de un conjunto de datos en relación con su media. Se utiliza para comparar la dispersión de diferentes conjuntos de datos que tienen diferentes unidades o escalas.

CV = (desvío estándar / media) * 100

El coeficiente de variación se expresa como un porcentaje y proporciona una medida de la variabilidad relativa en comparación con el tamaño de la media. 
`Cuanto mayor sea el coeficiente de variación, mayor será la variabilidad relativa de los datos en relación con su media. Por otro lado, cuanto menor sea el coeficiente de variación, menor será la variabilidad relativa de los datos en relación con su media.`

```python
## Coeficiente de variación

[np.std(equipo_a)/np.mean(equipo_a),
np.std(equipo_b)/np.mean(equipo_b)]
```
