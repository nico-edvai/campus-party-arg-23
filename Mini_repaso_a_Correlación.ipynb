 Juan decidió llevar a cabo un experimento. Registró la cantidad de minutos que dedicaba a practicar con modelos en un día y anotó la cantidad de desafíos que lograba superar ese mismo día. Registró estos datos durante varias semanas y luego decidió analizarlos.

Después de recopilar los datos, Juan utilizó los conceptos de correlación y coeficiente de determinación para entender mejor la relación entre los minutos de práctica y los desafíos superados. Utilizó la biblioteca NumPy en Python para realizar los cálculos necesarios.

Juan calculó la correlación entre los minutos de práctica y los desafíos superados y obtuvo un valor de correlación positivo. Esto significaba que a medida que Juan dedicaba más minutos a practicar con modelos, tenía más probabilidades de superar más desafíos. El coeficiente de correlación fue de aproximadamente 0.95, lo que indicaba una fuerte relación entre las dos variables.

Pero Juan no se detuvo ahí. También calculó el coeficiente de determinación, que proporciona información sobre la proporción de la variabilidad en los desafíos superados que se puede explicar por los minutos de práctica. Descubrió que el coeficiente de determinación era de aproximadamente 0.90, lo que implicaba que alrededor del 90% de la variabilidad en los desafíos superados se podía atribuir a los minutos de práctica en modelos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo: Relación entre minutos de práctica en modelos y desafíos superados
x = np.array([120, 191, 254, 290, 310, 428])  # Minutos de práctica en modelos
y = np.array([10, 15, 20, 25, 30, 35])  # Desafíos superados

# Cálculo de la correlación
corr_matrix = np.corrcoef(x, y)
corr = corr_matrix[0, 1]  # Valor de correlación entre x e y

# Cálculo del coeficiente de determinación
r_squared = corr**2

# Imprimir resultados
print("Coeficiente de correlación (R):", corr)
print("Coeficiente de determinación (R^2):", r_squared)

# Gráfico de dispersión con línea recta
plt.scatter(x, y)
plt.xlabel('Minutos de práctica en modelos')
plt.ylabel('Desafíos superados')
plt.title('Relación entre minutos de práctica en modelos y desafíos superados')

# Añadir línea recta
plt.plot(x, np.poly1d(np.polyfit(x, y, 1))(x), color='red')

plt.show()
```

El coeficiente de determinación (R^2) en relación con los resultados obtenidos por Juan.

El coeficiente de determinación, representado como R^2, es una medida estadística que proporciona información sobre la proporción de la variabilidad en la variable dependiente (en este caso, los desafíos superados) que puede ser explicada por la variable independiente (los minutos de práctica en modelos). En otras palabras, R^2 indica qué tan bien los valores de la variable independiente pueden predecir los valores de la variable dependiente.

En el caso de Juan, el coeficiente de determinación (R^2) fue de aproximadamente 0.90. Esto significa que alrededor del 90% de la variabilidad en los desafíos superados se puede atribuir a los minutos de práctica en modelos. En términos más sencillos, el 90% de las diferencias observadas en los desafíos superados por Juan pueden ser explicadas por la cantidad de tiempo que dedica a la práctica.

Un R^2 de 0.90 indica una relación muy fuerte entre los minutos de práctica y los desafíos superados. Esto significa que, en general, cuanto más tiempo dedica Juan a la práctica en modelos, más desafíos logra superar. Sin embargo, es importante tener en cuenta que el 10% restante de la variabilidad en los desafíos superados puede ser influenciado por otros factores que no se han tenido en cuenta en este estudio.

En resumen, el coeficiente de determinación (R^2) de 0.90 obtenido por Juan indica que la cantidad de tiempo que dedica a la práctica en modelos explica en gran medida su éxito en la superación de desafíos. Esta información respalda la importancia de la dedicación y el esfuerzo constante para alcanzar metas y desarrollar habilidades en el ámbito de los modelos matemáticos.

```python
import matplotlib.pyplot as plt

data = {
    'x1': [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0],
    'y1': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68],
    'x2': [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0],
    'y2': [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74],
    'x3': [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0],
    'y3': [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73],
    'x4': [8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0],
    'y5': [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
}


```

```python
# Crear una figura y subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Scatter plot 1
axs[0, 0].scatter(data['x1'], data['y1'], color='red')
axs[0, 0].set_title('Scatter Plot 1')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

# Ajustar línea y calcular R^2
coefficients = np.polyfit(data['x1'], data['y1'], 1)
line = np.polyval(coefficients, data['x1'])
residuals = data['y1'] - line
ss_res = np.sum(residuals**2)
ss_tot = np.sum((data['y1'] - np.mean(data['y1']))**2)
r_squared = 1 - (ss_res / ss_tot)

# Graficar línea
axs[0, 0].plot(data['x1'], line, color='black')

# Agregar R^2 en el gráfico
axs[0, 0].text(0.05, 0.95, f'R^2 = {r_squared:.2f}', transform=axs[0, 0].transAxes)

# Scatter plot 2
axs[0, 1].scatter(data['x2'], data['y2'], color='blue')
axs[0, 1].set_title('Scatter Plot 2')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Ajustar línea y calcular R^2
coefficients = np.polyfit(data['x2'], data['y2'], 1)
line = np.polyval(coefficients, data['x2'])
residuals = data['y2'] - line
ss_res = np.sum(residuals**2)
ss_tot = np.sum((data['y2'] - np.mean(data['y2']))**2)
r_squared = 1 - (ss_res / ss_tot)

# Graficar línea
axs[0, 1].plot(data['x2'], line, color='black')

# Agregar R^2 en el gráfico
axs[0, 1].text(0.05, 0.95, f'R^2 = {r_squared:.2f}', transform=axs[0, 1].transAxes)

# Scatter plot 3
axs[1, 0].scatter(data['x3'], data['y3'], color='green')
axs[1, 0].set_title('Scatter Plot 3')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# Ajustar línea y calcular R^2
coefficients = np.polyfit(data['x3'], data['y3'], 1)
line = np.polyval(coefficients, data['x3'])
residuals = data['y3'] - line
ss_res = np.sum(residuals**2)
ss_tot = np.sum((data['y3'] - np.mean(data['y3']))**2)
r_squared = 1 - (ss_res / ss_tot)

# Graficar línea
axs[1, 0].plot(data['x3'], line, color='black')

# Agregar R^2 en el gráfico
axs[1, 0].text(0.05, 0.95, f'R^2 = {r_squared:.2f}', transform=axs[1, 0].transAxes)

# Scatter plot 4
axs[1, 1].scatter(data['x4'], data['y5'], color='orange')
axs[1, 1].set_title('Scatter Plot 4')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

# Ajustar línea y calcular R^2
coefficients = np.polyfit(data['x4'], data['y5'], 1)
line = np.polyval(coefficients, data['x4'])
residuals = data['y5'] - line
ss_res = np.sum(residuals**2)
ss_tot = np.sum((data['y5'] - np.mean(data['y5']))**2)
r_squared = 1 - (ss_res / ss_tot)

# Graficar línea
axs[1, 1].plot(data['x4'], line, color='black')

# Agregar R^2 en el gráfico
axs[1, 1].text(0.05, 0.95, f'R^2 = {r_squared:.2f}', transform=axs[1, 1].transAxes)

# Ajustar la disposición de los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()

```
