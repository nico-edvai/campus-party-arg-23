48 atributos reales continuos [0,100] del tipo **word_freq_WORD**: Este atributo representa el porcentaje de palabras en el correo electrónico que coinciden con la PALABRA. Se calcula como 100 * (número de veces que la PALABRA aparece en el correo electrónico) / número total de palabras en el correo electrónico. Una "palabra" en este caso es cualquier cadena de caracteres alfanuméricos limitada por caracteres no alfanuméricos o fin de cadena.

Ejemplo: Si un correo electrónico tiene 100 palabras en total y la palabra 'descuento' aparece 10 veces, entonces el valor de 'word_freq_descuento' sería 10.

6 atributos reales continuos [0,100] del tipo **char_freq_CHAR**: Este atributo representa el porcentaje de caracteres en el correo electrónico que coinciden con el CARACTER. Se calcula como 100 * (número de veces que el CARACTER aparece) / total de caracteres en el correo electrónico.

Ejemplo: Si un correo electrónico tiene 1000 caracteres en total y el carácter '@' aparece 50 veces, entonces el valor de 'char_freq_@' sería 5.

1 atributo real continuo [1,...] del tipo **capital_run_length_average**: Este atributo representa la longitud promedio de secuencias ininterrumpidas de letras mayúsculas.

Ejemplo: Si un correo electrónico tiene secuencias de letras mayúsculas de longitudes 2, 4 y 6, el valor de 'capital_run_length_average' sería 4.

1 atributo entero continuo [1,...] del tipo **capital_run_length_longest**: Este atributo representa la longitud de la secuencia ininterrumpida más larga de letras mayúsculas.

Ejemplo: Continuando con el ejemplo anterior, el valor de 'capital_run_length_longest' sería 6.

1 atributo entero continuo [1,...] del tipo **capital_run_length_total**: Este atributo representa la suma de las longitudes de las secuencias ininterrumpidas de letras mayúsculas, o dicho de otra manera, el número total de letras mayúsculas en el correo electrónico.

Ejemplo: Continuando con el ejemplo anterior, el valor de 'capital_run_length_total' sería 2 + 4 + 6 = 12.

1 atributo nominal {0,1} del tipo **spam**: Este atributo indica si el correo electrónico fue considerado spam (1) o no (0).
